```python
# Functions accepts string, delimiter and nth occurence of the delimiter where string is split up. 
def split_at(s, delim, n):
    r = s.split(delim, n)[n]
    return s[:-len(r)-len(delim)], r

# Functions takes in a list of strings and split it up to return a list of strings
def get_country_info_list(testString):
    country_info_list = []
    for i in range(10): # There are 11 locations that start with 'A'. The for loop accounts for the first 10. 
        country_info_tuple = split_at(testString, ',', 4) # Splits the string on every 4th occurence of comma in the given string
        country_info_list.append(country_info_tuple[0])
        testString = country_info_tuple[1]
    country_info_list.append(testString) # This accounts for the last, remaining location.
    return country_info_list

# Function takes in a list of strings then splits each element into list of three elements ('urn:...', 'name:...', and 'facetUrn:...')
def get_country_info_final_form(copy_country_info_list):
    inner = []
    for country_info in copy_country_info_list:
        inner.append(country_info.split(','))     
    return inner

# Convert tuple into string
def convertTuple(tup):
	str = ''.join(tup)
	return str

# Function extracts key - value pairs, such as 'urn' : 'urn:...' 
def get_list_of_dict_separated_country_info(list_separated_country_info):
    outer = []
    for list_sep_info in list_separated_country_info: 
        dct = {}
        for attribute in list_sep_info:
            key = split_at(attribute, ':', 1)[0]
            key = key.replace('(', '') # remove the extra open bracket e.g. '(urn' becomes 'urn'
            value = split_at(attribute, ':', 1)[1:]
            value = convertTuple(value) # converting value from tuple to string removes the extra unwated characters like ','
            dct[key] = value
        # print(dct)
        outer.append(dct)
    return outer

# Extracts country names from each element in country info list and returns a list of the country names 
def get_country_names_list(copy_country_info_list):
    country_names = []
    for country_info in copy_country_info_list: 
        country_name = country_info.partition('name:')[2].split(',')[0] # get 
        country_names.append(country_name)
    return country_names

def makeDictionary(testString): 
    copy_country_info_list = get_country_info_list(testString)
    list_separated_country_info = get_country_info_final_form(copy_country_info_list)
    outer = get_list_of_dict_separated_country_info(list_separated_country_info)
    country_names_list = get_country_names_list(copy_country_info_list)
    # Create the locationsegments dict - for now, for countries that start with 'A'
    locationsegments = dict()
    # Add each country as a key and each country info as value 
    for i in range(len(country_names_list)):
        key = country_names_list[i]
        value = outer[i]
        locationsegments[key] = value
    return locationsegments
```
